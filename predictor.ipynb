```python
ordinal_map = {
    "Never smoked": 0,
    "Former smoker": 1,
    "Current smoker - now smokes some days": 2,
    "Current smoker - now smokes every day": 3
}

df["SmokerStatus"] = df["SmokerStatus"].map(ordinal_map).astype("Int64")
```

## <div class="custom-section"><div class="custom-header">step 2.2 = corrilation matrix</div><div class="custom-content">define encoders function</div></div>



```python
def encode_age_category(value):
    if value == "Age 18 to 24":
        low, high = 18, 24
    elif value == "Age 25 to 29":
        low, high = 25, 29
    elif value == "Age 30 to 34":
        low, high = 30, 34
    elif value == "Age 35 to 39":
        low, high = 35, 39
    elif value == "Age 40 to 44":
        low, high = 40, 44
    elif value == "Age 45 to 49":
        low, high = 45, 49
    elif value == "Age 50 to 54":
        low, high = 50, 54
    elif value == "Age 55 to 59":
        low, high = 55, 59
    elif value == "Age 60 to 64":
        low, high = 60, 64
    elif value == "Age 65 to 69":
        low, high = 65, 69
    elif value == "Age 70 to 74":
        low, high = 70, 74
    elif value == "Age 75 to 79":
        low, high = 75, 79
    elif value == "Age 80 or older":
        low, high = 80, 90
    else:
        return np.nan

    mean = (low + high) / 2
    std = (high - low) / 6

    sampled = np.random.normal(mean, std)
    return int(np.clip(round(sampled), low, high))


np.random.seed(42)

df["AgeCategory"] = df["AgeCategory"].apply(encode_age_category).astype("Int64")
```

```python
def encode_removed_teeth(value):
    if value == "None of them":
        low, high = 0, 0
    elif value == "1 to 5":
        low, high = 1, 5
    elif value == "6 or more, but not all":
        low, high = 6, 27
    elif value == "All":
        low, high = 28, 28
    else:
        return np.nan

    if low == high:
        return low

    mean = (low + high) / 2
    std = (high - low) / 6

    sampled = np.random.normal(mean, std)
    return int(np.clip(round(sampled), low, high))


np.random.seed(42)

df["RemovedTeeth"] = (
    df["RemovedTeeth"]
    .apply(encode_removed_teeth)
    .astype("Int64")
)
```

```python
def encode_diabetes(val):
    mapping = {
        "No": 0,
        "No, pre-diabetes or borderline diabetes": 1,
        "Yes, but only during pregnancy (female)": 2,
        "Yes": 3
    }
    return mapping.get(val, np.nan)

df['HadDiabetes'] = df['HadDiabetes'].map(encode_diabetes).astype("Int64")
```

```python
freq = df['State'].value_counts()
df['State'] = df['State'].map(freq)
```

```python
def encode_yes_no(df):
    yes_no_cols = []

    for col in df.columns:
        uniques = df[col].dropna().unique()
        if set(uniques) == {"Yes", "No"}:
            yes_no_cols.append(col)

    df[yes_no_cols] = df[yes_no_cols].replace({"Yes": 1, "No": 0}).astype("Int64")

    return df, yes_no_cols

df, yes_no_columns = encode_yes_no(df)

print("Yes/No columns:")
print(yes_no_columns)
```

```python
def encode_ast_checkup(value):
    if value == "Within past year (anytime less than 12 months ago)":
        low, high = 0, 12
    elif value == "Within past 2 years (1 year but less than 2 years ago)":
        low, high = 12, 24
    elif value == "Within past 5 years (2 years but less than 5 years ago)":
        low, high = 24, 60
    elif value == "5 or more years ago":
        low, high = 60, 120
    else:
        return np.nan

    mean = (low + high) / 2
    std = (high - low) / 6

    value = np.random.normal(mean, std)
    return np.clip(value, low, high)

df['LastCheckupTime'] = df['LastCheckupTime'].apply(encode_ast_checkup).round()

```

## <div class="custom-section"><div class="custom-header">Step 3.2 | fit model</div><div class="custom-content">RandomForestClassifier(optimize with optuna)</div></div>


```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

best_params = {
    'n_estimators': 254,
    'max_depth': 11,
    'min_samples_split': 14,
    'min_samples_leaf': 13
}

model = RandomForestClassifier(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=42
)

model.fit(X_train, y_train)
```

## <div class="custom-section"><div class="custom-header">Step 3.3 | metrics</div><div class="custom-content">confusion_matrix & report</div></div>


```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

y_pred = model.predict(X_test)
y_pred_tr = model.predict(X_train)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on Test Data: {accuracy * 100:.2f}%")


accuracy_tr = accuracy_score(y_train, y_pred_tr)
print(f"Accuracy on train Data: {accuracy_tr * 100:.2f}%")

print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```
